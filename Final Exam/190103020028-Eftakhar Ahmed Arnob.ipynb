## Ans of question : 2

```python
import pandas as pd
import numpy as np
import random


A = []
for i in range(0,100) :
    n = random.uniform(0,1)
    A.append(n)
    
B = []
for i in range(0,100) :
    n = random.uniform(0,1)
    B.append(n)
    
C = []
for i in range(0,100) :
    n = random.uniform(0,1)
    C.append(n)
    
D = []
for i in range(0,100) :
    n = random.uniform(0,1)
    D.append(n)
    
T = []
for i in range(0,100) :
    for i in range(0,100):
        if(A[i] + B[i] + C[i] + D[i] >= 0.5) :
            n = 1
        else:
            n = 0
    T.append(n)    


def generatedData() :
    data = pd.DataFrame({"A":A, "B":B, "C":C, "D":D, "T":T})
    print(data)

if __name__ == "__main__" :
    generatedData()
```
